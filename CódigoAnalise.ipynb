```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Carregar os dados
df = pd.read_csv('netflix_titles.csv')

# 2. Filtrar apenas o Brasil
filmes_BR = df[df['country'] == 'Brazil']

# 3. Contar quantos títulos existem por ano e colocar em ordem
contagem_por_ano = filmes_BR['release_year'].value_counts().sort_index()

# 4. Criar o gráfico de linhas
contagem_por_ano.plot(kind='line', marker='o', color='green', figsize=(10, 5))

# 5. Personalizar o gráfico
plt.title('Evolução de Lançamentos Brasileiros na Netflix 🇧🇷')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Quantidade de Títulos')
plt.grid(True)

# 6. Mostrar o gráfico
plt.show()
```
